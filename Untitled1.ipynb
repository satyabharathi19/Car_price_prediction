```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
import pickle

# Assuming 'f_df' is your DataFrame
sx = f_df.drop('Selling_Price', axis=1)  # Features
sy = f_df['Selling_Price']                # Target variable

sx_train, sx_test, sy_train, sy_test = train_test_split(sx, sy, test_size=0.2, random_state=42)

# Standardize the data
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(sx_train)
x_test_scaled = scaler.transform(sx_test)

# Using SVM model
svr = svm.SVR(kernel='linear')
svr.fit(x_train_scaled, sy_train)

# Make predictions
s_pred = svr.predict(x_test_scaled)
print(s_pred)

# Saving the model using pickle
with open('svm2_model.pkl', 'wb') as file:
    pickle.dump(svr, file)

# Loading the model
with open('svm2_model.pkl', 'rb') as file:
    svr = pickle.load(file)

print(type(svr))

# New data for prediction
sX_new = pd.DataFrame({
    'Present_Price': [5.59],
    'Kms_Driven': [27000],
    'Owner': [0],
    'no_year': [10],
    'Fuel_Type_Diesel': [0],
    'Fuel_Type_Petrol': [1],
    'Seller_Type_Individual': [0],
    'Transmission_Manual': [1],
    # Ensure order and features match training set
}, columns=sx.columns)

# Scale the new data
sX_new_scaled = scaler.transform(sX_new)

# Predict using the loaded model
s_predictions = svr.predict(sX_new_scaled)

# Display predictions
print("Predicted Selling Price:", s_predictions)
```
